# Integrity of the data source
The integrity of the data can be ensure by analyzing the data for bias and credibility, selecting the good data source and verifying data ethics, privacy and access.

**1. Bias and credibility**: This data source isn't a representative of the whole population. Suspecting data bias and sampling bias. Only Fitbit users data is available, not considering other smart device users.

**2. Good or bad data**: "The more high quality data we have, the more confidence we can have in our decisions". ROCCC process can be used to identify the good data source. ROCCC stands for **R**eliable, **O**riginal, **C**omprehensive, **C**urrent and **C**ited.

* This data source is **less reliable** because there are some issues with bias, accuracy and completeness of the data. 
* This data is **not Original**, it is a preprocessed version, and third-party data and need to be validated with the original source.
* This data is **some what Comprehensive**, roughly 70% of the critical informations needed to solve the business problem is available in this data.
* According to the scenario of this case study I am a Junior Data Analyst conducting this case study in 2016, so I could say this data **is current**.
* This data source is **not cited**.

**3. Licensing, Privacy, Security, and Accessibility**: The data has given the right for free access, usage and sharing since it is licensed under CC0: Public Domain and is open source. Privacy of the users were protected, all the personal identifiable informations(PII) is excluded in this data source.

### Query to join tables daily activity and daily sleep and saving as a temporary table 

WITH activity_sleep_merged as(

**SELECT**
    activity.Id,
    activity.Date,
    activity.Month,
    activity.Day,
    activity.TotalSteps,
    activity.TotalDistance,
    activity.Calories as CaloriesBurned,

    activity.VeryActiveDistance,
    activity.ModeratelyActiveDistance,
    activity.LightActiveDistance,
    activity.SedentaryActiveDistance,

    activity.VeryActiveMinutes,
    activity.FairlyActiveMinutes,
    activity.LightlyActiveMinutes,
    activity.SedentaryMinutes,

    sleep.TotalSleepRecords,
    sleep.TotalMinutesAsleep,
    sleep.TotalTimeInBed,

 **FROM** 
    
    `data-analytics-project-332412.Bellabeat.daily_activity_fitbit2016` as activity

 **FULL OUTER JOIN** 
     `data-analytics-project-332412.Bellabeat.daily_sleep_fitbit2016` as sleep
 
 ON activity.id = sleep.id AND activity.Date = sleep.Date 

)

SELECT *

FROM activity_sleep_merged ;


The above query results a table that is saved to google drive as a CSV file and **renamed as daily_activity_sleep_merged_fitbit2016. The link to this file is provided in the data source section of this notebook.
